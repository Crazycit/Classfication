```python
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
#Getting the data
data= datasets.load_iris()
```

```python
type(data)
```

```python
data
```

```python
iris= data.data
names= data.feature_names
```

```python
x= pd.DataFrame(iris, columns= names )
x.shape
```

```python
y= pd.DataFrame(data.target, columns= ["Target"])
y.shape
```

```python
df= pd.concat([x, y], axis=1)
```

```python
df.corr()["Target"]
```

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
plt.scatter(df["sepal length (cm)"], df["sepal width (cm)"], c= df["Target"], alpha= 0.8)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
scaler= StandardScaler()
pca= PCA()
lda= LinearDiscriminantAnalysis()
model= KNeighborsClassifier()
```

```python
params= {
    'pca__n_components': [2,3],
    'model__n_neighbors': [1,2,3,4,5,6,7,8,9],
    "lda__n_components": [1,2,3],
    "scaler__with_mean": [True, False],
    "scaler__with_std": [True, False]
}
pipeline= Pipeline([("scaler", scaler), ("pca", pca), ("lda", lda), ("model", model)])
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid= GridSearchCV(pipeline, params)
grid.fit(x, y)
```

```python
grid.best_params_
```

```python
grid.best_score_
```
